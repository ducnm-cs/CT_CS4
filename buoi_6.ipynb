```python
arr = list(map(str, input().split()))
dic = {}
for st in arr:
    pos = 0
    for i in range(0, len(st)):
        if st[i] == ':':
            pos = i
    k = st[0 : pos]
    val = st[pos + 1 : len(st)]
    if k in dic:
        dic[k].append(val)
    else:
        dic[k] = []
        dic[k].append(val)
print(dic)
```

```python
d1 = eval(input())
d2 = eval(input())
res = {}
for k in d1.keys():
    if k in res:
        res[k] += d1[k]
    else:
        res[k] = d1[k]
for k in d2.keys():
    if k in res:
        res[k] += d2[k]
    else:
        res[k] = d2[k]
print(dict(sorted(res.items())))
```

```python
arr = list(map(int, input().split()))
k = int(input())
cnt = {}
res = []
used = {}
for a in arr:
    cnt[a] = 1
for a in cnt.keys():
    if ((k - a) in cnt) and ((a, k - a) not in used) and ((k - a, a) not in used):
        res.append((a, k - a))
        used[(a, k - a)] = used[(k - a, a)] = 1
print(res)
```

```python
arr = list(map(int, input().split()))
arr.sort()
cnt = {}
for i in arr:
    if i in cnt:
        cnt[i] += 1
    else:
        cnt[i] = 1
res = arr[0]
for k in cnt.keys():
    if cnt[k] > cnt[res]:
        res = k
print(res)
```

```python
n, m, k = map(int, input().split())
a = []
for i in range(0, n):
    a.append(list(map(int, input().split())))
rw = 0; cl = 0
if k <= n and k <= m:
    for i in range(0, m):
        rw += a[k - 1][i]
    for i in range(0, n):
        cl += a[i][k - 1]
    print(rw, cl)
elif n <= k <= m:
    for i in range(0, n):
        cl += a[i][k - 1]
    print(cl)
elif m <= k <= n:
    for i in range(0, m):
        rw += a[k - 1][i]
    print(rw)
```
