```python
#3
def happy_num(n):
    tmp = 0
    while n > 0:
        tmp += (n % 10) * (n % 10)
        n //= 10
        
    if tmp == 1:
        return True
    else:
        return happy_num(tmp)
n = int(input())
try:
    if happy_num(n) == True:
        print("YES")
except RecursionError:
    print("NO")
    
```
