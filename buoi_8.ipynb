```python
#1
class cylinder:
    def __init__(self, rad, he):
        self.radius = rad
        self.height = he
    def surface_area(self):
        print("{:.2f}".format(self.radius * self.radius * 3.14 * 2 + self.height *  (self.radius * 2 * 3.14)))
    def vol(self):
        print("{:.2f}".format(self.radius * self.radius * 3.14 * 2 * self.height))
a, b = map(int, input().split())
tmp = cylinder(a, b)
tmp.surface_area()
tmp.vol()
```

```python
#2
class Date:
    def __init__(self, d, m, y):
        self.day = d
        self.month = m
        self.year = y
    def next_day(self):
        check = 1
        if self.month == 2:
            if self.year % 4 == 0 and self.year % 100 != 0:
                if self.day > 29:
                    check = 0
            elif self.day > 28:
                check = 0
        elif self.month % 2 == 1 and self.month <= 7:
            if self.day > 31:
                check = 0
        elif self.month % 2 == 0 and self.month <= 7:
            if self.day > 30:
                check = 0
        elif self.month % 2 == 0 and self.month > 7:
            if self.day > 31:
                check = 0
        elif self.month % 2 == 1 and self.month > 7:
            if self.day > 30:
                check = 0
        if check == 0:
            print("INVALID")
        else:
            if self.month == 1 or self.month == 3 or self.month == 5 or self.month == 7 or self.month == 8 or self.month == 10:
                if self.day == 31:
                    print(f"1/{self.month + 1}/{self.year}")
                else:
                    print(f"{self.day + 1}/{self.month}/{self.year}")
            elif self.month == 2:
                if self.year % 4 == 0 and self.year % 100 != 0:
                    if self.day == 29:
                        print(f"1/3/{self.year}")
                    else:
                        print(f"{self.day + 1}/2/{self.year}")
                else:
                    if self.day == 28:
                        print(f"1/3/{self.year}")
                    else:
                        print(f"{self.day + 1}/2/{self.year}")
            elif self.month == 12:
                if self.day == 31:
                    print(f"1/1/{self.year + 1}")
                else:
                    print(f"{self.day + 1}/12/{self.year}")
            else:
                if self.day == 30:
                    print(f"1/{self.month + 1}/{self.year}")
                else:
                    print(f"{self.day + 1}/{self.month}/{self.year}")
d, m, y = map(int, input().split('/'))
tmp = Date(d, m, y)
tmp.next_day()
```

```python
#3
from math import sqrt
class Coor:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def dist(self, self2):
        print("{:.2f}".format(sqrt((self.x - self2.x) ** 2 + (self.y - self2.y) ** 2)))
    def check(self):
        if self.x == 0 and self.y == 0:
            print("La goc toa do")
        if self.x == 0:
            print("Nam tren truc tung")
        if self.y == 0:
            print("Nam tren truc hoanh")
x, y = map(int, input().split())
tmp1 = Coor(x, y)
tmp2 = Coor(0, 0)
tmp1.check()
tmp1.dist(tmp2)
```

```python
#6
class rectangle:
    def __init__(self, w, h):
        self.width = min(w, h)
        self.length = max(w, h)
    def parameter(self):
        return (self.width + self.length) * 2
    def area(self):
        return self.width * self.length
    def scale(self, mul):
        self.width *= mul
        self.length *= mul
w, l, k = map(int, input().split())
tmp = rectangle(w, l)
tmp.scale(k)
print("{:.2f}".format(tmp.area()), end = ' ')
print("{:.2f}".format(tmp.parameter()))
```

```python
#7
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)
class Fraction:
    def __init__(self, a, b):
        self.tu = a
        self.mau = b
    def calc(self, self2, op):
        if op == '+':
            ts = self.tu * self2.mau + self2.tu * self.mau
            ms = self.mau * self2.mau
            uoc = gcd(ts, ms)
            print(f"{ts//uoc}/{ms//uoc}")
        if op == '-':
            ts = self.tu * self2.mau - self2.tu * self.mau
            ms = self.mau * self2.mau
            uoc = gcd(ts, ms)
            print(f"{ts//uoc}/{ms//uoc}")
        if op == '*':
            ts = self.tu * self2.tu
            ms = self.mau * self2.mau
            uoc = gcd(ts, ms)
            print(f"{ts//uoc}/{ms//uoc}")
        if op == '/':
            ts = self.tu * self2.mau
            ms = self.mau * self2.tu
            uoc = gcd(ts, ms)
            print(f"{ts//uoc}/{ms//uoc}")
a, b, op, c, d = map(str, input().split())
tmp1 = Fraction(int(a), int(b))
tmp2 = Fraction(int(c), int(d))
tmp1.calc(tmp2, op)
```

```python
#8
class BankAccount:
    def __init__(self, m):
        self.balance = m
    def access(self, cmd, m):
        if cmd == 'DEPOSIT':
            self.balance += m
        elif cmd == 'WITHDRAW':
            if self.balance >= m:
                self.balance -= m
    def check(self):
        print("{:.2f}".format(self.balance))
a, m = map(str, input().split())
tmp = BankAccount(float(m))
t = int(input())
while t > 0:
    t -= 1
    action, money = map(str, input().split())
    tmp.access(action, float(money))
tmp.check()
```

```python
#11
import numpy as np
class Vector:
    def __init__(self, x, y, z):
        self.vt = np.array([x, y, z])

    def cross_prod(self, self2):
        return np.cross(self.vt, self2.vt)
    
    def dot_prod(self, self2):
        return np.dot(self.vt, self2.vt)
    
    def magnitude(self):
        return np.linalg.norm(self.vt)
    
x1, y1, z1 = map(int, input().split()) #A
x2, y2, z2 = map(int, input().split()) #B
x3, y3, z3 = map(int, input().split()) #C
x4, y4, z4 = map(int, input().split()) #D

AB = Vector(x2-x1, y2-y1, z2-z1)
BC = Vector(x3-x2, y3-y2, z3-z2)
CD = Vector(x4-x3, y4-y3, z4-z3)

NV1 = AB.cross_prod(BC)
NV2 = BC.cross_prod(CD)

ts = np.dot(NV1, NV2)
ms = np.linalg.norm(NV1) * np.linalg.norm(NV2)

cos = ts / ms

res = np.degrees(np.arccos(cos))

print("{:.2f}".format(res))
```

```python
#12
class ImaginaryNumber:
    def __init__(self, r, i):
        self.re = r
        self.im = i

    def multiplication(self, self2):
        real = self.re * self2.re - self.im * self2.im
        img = self.re * self2.im + self.im * self2.re
        print(f"{real} {img}")

a, b = map(int, input().split())
c, d = map(int, input().split())
tmp1 = ImaginaryNumber(a, b)
tmp2 = ImaginaryNumber(c, d)

tmp1.multiplication(tmp2)
```

```python
#13
from math import sqrt
class Triangle:
    def __init__(self, a, b, c):
        self.x = a
        self.y = b
        self.z = c

    def area(self):
        if self.x + self.y > self.z and self.x + self.z > self.y and self.y + self.z > self.x:
            half_par = (self.x + self.y + self.z) / 2
            res = sqrt(half_par * (half_par - self.x) * (half_par - self.y) * (half_par -self.z))
            print("{:.2f}".format(res))
        else:
            print("invalid")
a, b, c = map(int, input().split())
tmp = Triangle(a, b, c)
tmp.area()
```
