```python
#1
def binsearch(arr, l, r, n):
    mid = int(l + (r - l) / 2)
    if l > r:
        return -1
    if arr[mid] == n:
        return mid
    elif arr[mid] < n:
        return binsearch(arr, mid + 1, r, n)
    else:
        return binsearch(arr, l, mid - 1, n)
arr = list(map(int, input().split()))
n = int(input())
print(binsearch(arr, 0, len(arr) - 1, n))
```

```python
#2
def count_occur(arr, n):
    cnt = 0
    for i in arr:
        if i == n:
            cnt += 1
    return cnt
arr = list(map(int, input().split()))
n = int(input())
print(count_occur(arr, n))
```

```python
#3
def sort_check(arr):
    for i in range(0, len(arr) - 1):
        if arr[i] > arr[i + 1]:
            return False
    return True
def bb_sort(arr):
    while sort_check(arr) == False:
        for i in range(0, len(arr) - 1):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]
    return arr
arr = list(map(int, input().split()))
new_arr = bb_sort(arr)
for i in new_arr:
    print(i, end = ' ')
```

```python
#7
a = eval(input())
pos = [0] * 1000001
pos[1] = 0
res = 1
def binsearch(l, r, val):
    res = 0
    while l <= r:
        mid = (l + r) // 2
        if a[pos[mid]] < val:
            res = mid
            l = mid + 1
        else:
            r = mid - 1
    return res

for i in range(1, len(a)):
    tmp = binsearch(1, res, a[i])
    if tmp + 1 > res:
        res = tmp + 1
        pos[tmp + 1] = i
    elif a[i] < a[pos[tmp + 1]]:
        pos[tmp + 1] = i
print(res)
```

```python
#8
a = eval(input())
q = eval(input())
res = [-1] * len(q)
q.sort()
def binsearch_left(l, r, x):
    res = 0
    while l <= r and r < len(q) and l >= 0:
        mid = (l + r) // 2
        if q[mid] >= x:
            res = mid
            r = mid - 1
        else:
            l = mid + 1
    return res

def binsearch_right(l, r, x):
    res = 0
    while l <= r and r < len(q) and l >= 0:
        mid = (l + r) // 2
        if q[mid] <= x:
            res = mid
            l = mid + 1
        else:
            r = mid - 1
    return res
    
for inv in a:
    tmp_l = binsearch_left(0, len(q) - 1, inv[0])
    tmp_r = binsearch_right(0, len(q) - 1, inv[1])
    for i in range(tmp_l, min(len(q), tmp_r + 1)):
        if res[i] == -1:
            res[i] = inv[1] - inv[0]
        else:
            res[i] = min(res[i], inv[1] - inv[0])

print(res)
```

```python
#9
a = eval(input())
cnt = {}
for i in a:
    cnt[i] = 1
sort_map = dict(sorted(cnt.items()))
dem = 1
for k in sort_map.keys():
    sort_map[k] = dem
    dem += 1

for i in range(0, len(a)):
    a[i] = sort_map[a[i]]

bit = [0] * (len(a) + 1)
def getsum(idx):
    res = 0
    while idx > 0:
        res += bit[idx]
        idx -= (idx & (-idx))
    return res

def update(u, val):
    idx = u
    while idx < len(a) + 1:
        bit[idx] += val
        idx += (idx & (-idx))

res = []
for i in range(len(a) - 1, -1, -1):
    #print(i)
    update(a[i], 1)
    res.append(getsum(a[i] - 1))
res.reverse()
print(res)
```
