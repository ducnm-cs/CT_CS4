```python
def swap(a, b):
    a, b = b, a
    return (a, b)
a, b = map(int, input().split())
tmp = swap(a, b)
print(tmp[0], tmp[1])
```

```python
def snt(n):
    idx = 2
    while idx * idx <= n:
        if n % idx == 0:
            return False
        idx += 1
    return n >= 2
a = int(input())
print(snt(a))
```

```python
def shh(n):
    res = 1
    idx = 2;
    while idx * idx <= n:
        if n % idx == 0:
            res += idx
            if idx != n / idx:
                res += n / idx
        idx += 1
    return res == n
a = int(input())
print(shh(a))
```

```python
def exec(arr, n):
    idx = 1
    for a in arr:
        if a == n:
            return idx
        idx += 1
    return -1
arr = list(map(int, input().split()))
n = int(input())
print(exec(arr, n))
```

```python
def gt(n):
    res = 1
    for i in range(1, n + 1):
        res *= i
    return res
n = int(input())
print(gt(n))
```

```python
a, b = map(int, input().split())
tmp = a ^ b
idx = 0
res = 0
while 1 << idx <= tmp:
    if tmp & (1 << idx) != 0:
        res += 1
    idx += 1
print(res)
```

```python
def sumdigit(n):
    sum = 0
    while n > 0:
        sum += n % 10
        n //= 10
    return sum
n = int(input())
print(sumdigit(n))
```
